```python
import pandas as pd

df = pd.read_csv('/content/dataset_vendas.csv')
```

#Diagnóstico do Dataset

```python
# visualizando as informações do dataset

df.info()
```

```python
# Verificando se há valores nulos

df.isnull().sum()
```

```python
# verificando se há outliers

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

if not numeric_cols.empty:
    Q1 = df[numeric_cols].quantile(0.25)
    Q3 = df[numeric_cols].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = []

    for col in numeric_cols:
        df = df[(df[col] >= limite_inferior[col]) & (df[col] <= limite_superior[col])]

    print("Outliers encontrados:", outliers)
    print("Quantidade de outliers:", len(outliers))
else:
    print("Nenhuma coluna numérica encontrada para tratamento de outliers.")
```

# Tratamento dos dados

```python
# removendo duplicados

df.drop_duplicates(inplace=True)
```

```python
# criação da coluna mês
df['data'] = pd.to_datetime(df['data'])
df['mes'] = df['data'].dt.month
```

```python
# criação da coluna ticket médio

df['ticket_medio'] = df['faturamento'] / df['quantidade']
```

```python
# salvando as alterações em csv

df.to_csv('dataset_vendas_tratado.csv', index=False)
```
